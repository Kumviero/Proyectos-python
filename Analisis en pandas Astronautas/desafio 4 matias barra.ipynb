### - Transformación y manipulación de columnas

Transformamos la altura de cm a metros para conseguir valores reales del IMC. y creamos la columna IMC.

```python
df["IMC"] = df["peso"] / ((df["altura"]/100)**2) # formula de IMC en caso de que altura este en CM.
```

```python
df.head()
```

Creamos columna Promedio Evaluación la cual tomara las evaluaciones fisicas, psicologicas, adaptativas y medicas de cada registro.

```python
cols_eval = ["evaluacion_fisica", "evaluacion_psicologica", "evaluacion_adaptativa", "evaluacion_medica"] # seleccionamos las columnas que ocuparemos
df["promedio_evaluaciones"] = df[cols_eval].mean(axis=1) #creamos columna promedio_evaluaciones con .mean que es la funcion de promedio
```

```python
df.head() #miramos de nuevo como salio la transformación
```

```python
df_filtrado = df[ ## creamos un df diferente para incluir las registros filtrados
    (df["IMC"].between(18.5, 24.9)) &
    (df["promedio_evaluaciones"] >= 87) #la condicion busca los IMc entre 18.5 y 24.9, para luego evaluar su promedio
]
```

```python
df_filtrado = df_filtrado.reset_index(drop=True) #reiniciamos los indices del Df ya que con el filtrado ya no se hace fiable su uso.
```

```python
df_filtrado["califica"] = "Si" # cambiamos los datos de columna califica a si, dado que deberia calificar.
```
