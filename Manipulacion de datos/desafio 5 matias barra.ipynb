## <span style="color: green;"> b.</span> Genera una tabla que contenga la unión de las 3 tablas officer, subjects y incidents. 

Primero, seleccionamos 2 df y las unimos en 1 solo merge, dadoque pandas no puede trabajar con mas de dos df a la vez, y utilizamos la columna clave que comparten nuestras df: case_number

```python
df_incidents_officers = pd.merge(
    df_incidents,                 # tabla base
    df_officers,                  # tabla a unir
    on="case_number",          # columna en comun/columna clave
    how="left",                # conserva todos los incidents, aunque no haya oficiales
    suffixes=("", "_officers") # agrega sufijo a columnas de officers que se repitan en incidents.
)
```

```python
df_incidents_officers.info()
```

 Ahora unimos el resultado con subjects (también por 'case_number')

```python
df_merged_data = pd.merge(
    df_incidents_officers,        # lo que ya unimos
    df_subjects,                  # tabla de sujetos
    on="case_number",          # columna comun/columna clave
    how="left",                 # mismo tipo de join
    suffixes=("", "_subject")  # agrega sufijo a columnas de subjects que se repitan en incidents_officers
)
```

Verificamos que el df_merged_data se puede visualizar.

```python
df_merged_data.info()
```

## <span style="color: green;"> c.</span> Verifica si hay filas duplicadas; si es así, elimínalas

Contamos cuantas filas duplicadas existen en el DataFrame unido

```python
duplicados = df_merged_data.duplicated().sum()
print(f"Filas duplicadas encontradas: {duplicados}")
```

no hay duplicadas en si (dado que el registro tiene 24 columnas y es dificil que se repita el mismo registro, lo que podemos hacer es condicionar que el conteo de los duplicados se enfoque en columnas especificas)

```python
duplicados = df_merged_data.duplicated(subset=["case_number", "full_name", "full_name_subject"]).sum() ## buscamos registros cuyo case_number, full_name y full_name_subject sean igual
print(f"Filas duplicadas encontradas: {duplicados}")
```

Ahora eliminamos filas duplicadas, en si comparamos que el numero de caso en cuestion no tenga el mismo nombre asociado tanto en officer como en subject.

```python
df_merged_data = df_merged_data.drop_duplicates(subset=["case_number", "full_name", "full_name_subject"])
```

## <span style="color: green;"> f.</span> Genera una tabla pivote que muestre en las filas el género del oficial y en las columnas el género del subject. ¿Cómo interpretas los valores que muestra esta vista? 

```python
tabla_pivote = pd.pivot_table(
    df_merged_data,
    index="gender",   # filas: género del oficial
    columns="gender_subject", # columnas: genero del sujeto
    values="case_number",     # cualquier valor (solo para contar)
    aggfunc="count",          # función de agregación → cuenta cuántos casos hay
    fill_value=0              # reemplaza NaN por 0
)
```

## <span style="color: green;"> f.</span> Finalmente, genera una tabla pivote que muestre la media del salario estimado mínimo y la media del salario estimado máximo por tamaño de empresa

```python
tabla_pivote_salarios = pd.pivot_table(
    jobs_reemplazado,
    index="Size",  # Filas -> tamaño de empresa
    values=["Salario Estimado Minimo", "Salario Estimado Maximo"],  # Qué calcular
    aggfunc="mean"  # Qué operación aplicar -> promedio
)
tabla_pivote_salarios.head()
```
